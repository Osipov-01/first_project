```python
input_file = open('input.txt', 'r')
data = [line.strip() for line in input_file]
```

```python
print(data)
```

```python
calories_total_list = []
calories_total = 0
for i in range(len(data)):    
    if data[i] == '':
        calories_total_list.append(calories_total)
        calories_total = 0
    else:
        calories = int(data[i])
        print(data[i])
        print(calories)
        print('')
        calories_total += calories
```

```python
print(calories_total_list)
```

```python
print(max(calories_total_list))
```

```python
max_three_calories_sum = 0
for i in range(0, 3):
    max_three_calories_sum += max(calories_total_list)
    calories_total_list.remove(max(calories_total_list))
    print(max(calories_total_list))
```

```python
print(max_three_calories_sum)
```
